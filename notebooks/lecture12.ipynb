## ☛ Hands-on exercise: matplotlib



-   Let's start with a simple plot and then modify it
-   Look at `legend`, `set_xticklabels`, `set_xticks`, `title`, `set_xlabel` etc.



```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
ax.plot(np.random.randn(30).cumsum(), 'ko--', label="Line 1")
```

-   Change the limits of the x-axis to [-10, 40]



```python
import matplotlib
matplotlib.__version__
```

```python
plt.xlim?
```

```python
ax.set_xlim((-10, 40))
fig
```

-   Add a legend in the top left corner



```python
ax.legend?
```

```python
ax.legend(loc='upper left')
fig
```

-   Change the limits of the x-axis back to [0,30] and set ticks to A, B, C at positions 5, 15, and 25



```python
ax.set_xlim([0, 30])
lbs = ax.get_xticklabels()
s = lbs[0]
s.set_label
```

-   Add a title to your plot



```python
ax.set_title("A title")
fig
```

-   Rotate your x-axis labels by 90 degrees



```python
ax.set_xticks([5, 15, 25])
ax.set_xticklabels(['A', 'B', 'C'], rotation=90)
fig
```

-   Change the font size of the y-axis labels



```python
ax.set_yticklabels(ax.get_yticklabels(), fontsize=20)
fig
```

### Boxplots



Let's load an example dataset



```python
tips = sns.load_dataset("tips")
tips.head()
```

Look at the documentation of `sns.boxplot` and draw a nested boxplot to show bills by day and time



```python
sns.boxplot?
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
sns.boxplot(data=tips, y='total_bill', x='day', hue='time', ax=ax)
```

### Scatter plots



Point plots are a useful way of examining the relationship between datasets



```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

Look at the documentation of `sns.scatterplot` and create a plot with carat size on the x-axis, price on the y-axis with points having different colors according to clarity and different sizes according to depth



```python
sns.scatterplot?
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', size='depth', ax=ax, linewidth=0)
```

Now go back to the `tips` dataset, and plot a scatterplot with the `total_bill` on the x-axis and `tip` on the y-axis, but this time add a regression line



```python
sns.lmplot(data=tips, x='total_bill', y='tip')
```

Can you set the regression confidence to 68%?



```python
sns.lmplot(data=tips, x='total_bill', y='tip', ci=68)
```

Now fit a 2nd order polynomial to the data and remove the confidence intervals



```python
sns.lmplot(data=tips, x='total_bill', y='tip', order=2, ci=None)
```

### Facet grids



-   We can visualize subsets of data grouping them on categorical data
-   Look at the `sns.catplot` function to create two plots side-by-side (left for lunch, right for dinner), with `day` on the x-axis, `tip` on the y-axis, different colors for `sex`



```python
sns.catplot(data=tips, x='day', y='tip', col='time', hue='sex')
```

-   We can also be a bit more complicated by drawing custom plots on each facet
-   Create a $2 \times 2$ set of plots with rows designating `sex` and columns `time`, and each subplot showing the histogram of the `total_bill`
-   Look at the `sns.FacetGrid` class



```python
g = sns.FacetGrid(data=tips, row='sex', col='time')
g.map(sns.histplot, 'total_bill')
```
